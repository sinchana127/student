# reduce using numpy

```python
import numpy as np
from functools import reduce
arr=np.array([1,2,3,4,5])
sum_arr=reduce(lambda x,y:x+y,arr)
mul_arr=reduce(lambda x,y:x*y,arr)
print(sum_arr)
print(mul_arr)
```

# lambda using numpy

```python
import numpy as np
arr=np.array([1,2,3,4,5])
squared_arr=np.array(list(map(lambda x:x**2,arr)))
doubled_arr=np.array(list(map(lambda x:x*3,arr)))
increment_arr=np.array(list(map(lambda x:x+1,arr)))
print(squared_arr)
print(doubled_arr)
print(increment_arr)
```

# histogram

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.DataFrame ({
    'A':[1,2,3,4,5],
    'B':[5,4,3,2,1],
    'C':[2,3,4,5,6]
})
df.plot(kind='line',title='line plot')
plt.show()
df.plot(kind='bar',title='bar plot')
plt.show()
df['A'].plot(kind='hist', bins=10,title='histogram plot')
plt.show()
df.plot(kind='box',title='box plot')
plt.show()
df.plot(kind='scatter',x='A',y='B',title='scatter plot')
plt.show()
fig,axes=plt.subplots(nrows=2,ncols=2,figsize=(12,10))
df['A'].plot(ax=axes[0,0],title=('subplot 1'))
df['B'].plot(ax=axes[0,0],title=('subplot 2'))
df['C'].plot(ax=axes[0,0],title=('subplot 3'))
df.plot(kind='scatter',x='A',y='C',ax=axes[1,1],title='subplot 4')
plt.tight_layout()
plt.show()

```
