```python
import pandas as pd
df=pd.read_csv("/content/StudentsPerformance.csv")
df["math score"]=df["math score"].apply(lambda x: x+2)
print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.scatter(df["math score"],df["reading score"])
plt.title("math score and reading score")
plt.xlabel("x-axis math score")
plt.ylabel("y-axis reading score")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.hist(df["math score"], bins=10)
plt.title("math score and reading score")
plt.xlabel("x-axis math score")
plt.ylabel("y-axis reading score")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv("/content/StudentsPerformance.csv")
sns.boxplot(x="math score", y="writing score", data=data)
plt.title("math score and writing score")
plt.show()
```
