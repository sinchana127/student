```python
#Univariate outlier detection
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("/content/Occupations.csv")
column='HoursPerWeek'
z_scores=np.abs((df['HoursPerWeek']-df['HoursPerWeek'].mean())/df['HoursPerWeek'].std())
outliers=df[z_scores>3]
print("Outliers")
print(outliers)
sns.boxplot(x=df['EducationYears'])
plt.title('Box Plot for Outlier Detection')
plt.show()
```

```python
#Bivariate outlier detection
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("/content/Occupations.csv")
numeric_df = df.select_dtypes(include=[np.number])
df['Age']=np.sqrt(((numeric_df-numeric_df.mean())/numeric_df.std()).pow(2).sum(axis=1))
outliers=df[df['Age']>3]
print(outliers)
plt.scatter(df['FinalWeight'], df['Age'])
plt.title('Scatter Plot for Bivariate Outlier Detection')
plt.xlabel('Finalweight')
plt.ylabel('Age')
plt.show()
```

```python
#TimeSeries outlier deletion
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv('/content/Occupations.csv')
data['Age'] = pd.to_datetime(data['Age'])
# Calculate rolling mean and standard deviation
rolling_mean = data['EducationYears'].rolling(window=30).mean()
rolling_std = data['EducationYears'].rolling(window=30).std()
outliers = np.abs(data['EducationYears'] - rolling_mean) > 2 * rolling_std
print("Time Series Outliers:")
print(data[outliers])
data=data[~outliers]
plt.plot(data.index, data['EducationYears'])
plt.xlabel('Day')
plt.ylabel('EducationYears')
plt.title('Time Series Outliers Deletion')
plt.show()
```

```python
#Numerosity data reduction
import pandas as pd
import numpy as np
df=pd.read_csv("/content/Occupations.csv")
def numerosity_reduction(df,precision):
  return np.round_(df, decimals = precision)
df['reduced']=numerosity_reduction(df['EducationYears'],2)
print(df)
plt.hist(df['EducationYears'], bins=20, alpha=0.5, label='Original')
plt.hist(df['reduced'], bins=20, alpha=0.5, label='Reduced')
plt.xlabel('Education Years')
plt.ylabel('Frequency')
plt.title('Distribution of Education Years Before and After Numerosity Reduction')
plt.show()
```

```python
#Principla component analysis
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt
df=pd.read_csv("/content/Occupations.csv")
numeric_df = df.select_dtypes(include=['number'])
imputer = SimpleImputer(strategy='mean')
numeric_df_imputed = imputer.fit_transform(numeric_df)
pca=PCA(n_components=2)
reduced_data=pca.fit_transform(numeric_df_imputed)
reduced_df=pd.DataFrame(reduced_data,columns=['PC1','PC2'])
print(reduced_df)
plt.scatter(reduced_df['PC1'],reduced_df['PC2'])
plt.xlabel('Contry_Code')
plt.ylabel('Year')
plt.title('PCA')
plt.show()
```

```python
#LDA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
iris=load_iris()
x=iris.data
y=iris.target
lda=LinearDiscriminantAnalysis(n_components=2)
x_lda=lda.fit_transform(x,y)
plt.scatter(x_lda[:,0],x_lda[:,1],c=y,cmap='viridis')
plt.xlabel('LDA Component 1')
plt.ylabel('LDA Component 2')
plt.title('Linear Discriminant Analysis')
plt.show()
```

```python
#t-SNE
import pandas as pd
from sklearn.manifold import TSNE
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt
df=pd.read_csv("/content/Occupations.csv")
numeric_df = df.select_dtypes(include=['number'])
imputer = SimpleImputer(strategy='mean')
numeric_df_imputed = imputer.fit_transform(numeric_df)
tsne = TSNE(n_components=2)
reduced_data=tsne.fit_transform(numeric_df_imputed)
reduced_df=pd.DataFrame(reduced_data,columns=['Dimension 1','Dimension 2'])
print(reduced_df)
plt.scatter(reduced_df['Dimension 1'],reduced_df['Dimension 2'])
plt.xlabel('Year')
plt.ylabel('Contry_Code')

plt.title('t-SNE Visualization')
plt.show()
```
