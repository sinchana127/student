```python
#T-test
import pandas as pd
from scipy import stats
data=pd.read_csv("/content/student marks.csv")
t_stat,p_val=stats.ttest_ind(data['Kannada'],data['English'])
print("T-statistic:",t_stat)
print("P-value:",p_val)

```

```python
#Chi -square
import pandas as pd
from scipy.stats import chi2_contingency
data=pd.read_csv("/content/student marks.csv")
contingency_table=pd.crosstab(data['Kannada'],data['English'])
chi2,p,dof,expected=chi2_contingency(contingency_table)
print("Chi-Square stastistic:",chi2)
print("P-value:",p)
print("degree of freedom:",dof)
```

```python
#covariance calculation
import numpy as np
def calculate_covariance(x, y):
    n = len(x)
    if n != len(y):
      raise ValueError("Datasets X and Y must have the same length")
    mean_x=np.mean(x)
    mean_y=np.mean(y)
    covariance = sum((x[i] - mean_x) * (y[i] - mean_y) for i in range(n)) / (n-1)
    return covariance
x=[1,2,3,4,5,6]
y=[2,4,7,9,12,14]
covarience=calculate_covariance(x,y)
print("Covarience:", covarience)
```

```python
#correlation calculation
def calculate_correlation(x,y):
  covariance=calculate_covariance(x,y)
  std_dev_x=np.std(x, ddof=1)
  std_dev_y=np.std(y, ddof=1)
  correlation=covariance / (std_dev_x * std_dev_y)
  return correlation
x=[1,2,3,4,5,6]
correlation=calculate_correlation(x,y)
print("Correlation:", correlation)
```
